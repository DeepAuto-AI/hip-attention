```python
markers = [
    int(line.split(":")[-1].strip()) for line in lines if line.startswith("marker")
]
```

```python
plt.hist(markers, bins=30)
plt.xlabel("load blocks")
plt.show()
print(sum(markers) / len(markers))
```
